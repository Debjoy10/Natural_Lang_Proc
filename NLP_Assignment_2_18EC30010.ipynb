```python
# read dataset
with open('corpus.txt', 'r') as f:
    corpus = f.read()
```

Preprocess the data

```python
import nltk
nltk.download('punkt') # For tokenizers
from nltk.tokenize import word_tokenize,sent_tokenize

# Preprocessing
import re
sents = sent_tokenize(corpus)
sents_proc = []

def sent_proc(s):
    # 1. Removing Punctuations
    res = re.sub(r'[^\w\s]', '', s) 
    # 2. Remove Alphanumeric
    res = re.sub(r'\w*\d\w*', '', res)
    # 3. Convert to Lowercase
    res = res.lower()
    # 4. Strip Whitespaces/Underscores
    res = res.replace('_', '')
    
    return word_tokenize(res)

for s in sents:
    res = sent_proc(s)
    if res is not []:
        sents_proc.append(res)
```

```python
# Write code

from nltk.util import ngrams
unigrams=[]
bigrams=[]
trigrams=[]
fourgrams=[]

for content in sents_proc:
    unigrams.extend(ngrams(content,1))
    bigrams.extend(ngrams(content,2))
    trigrams.extend(ngrams(content,3))
    fourgrams.extend(ngrams(content,4))
```

```python
#stopwords = code for downloading stop words through nltk
from nltk.corpus import stopwords
stopwords = stopwords.words('english')

#print top 10 unigrams, bigrams after removing stopwords
uni_fdist_sw = nltk.FreqDist(unigrams)
uni_processed = [p for p in unigrams if p[0] not in stopwords]
uni_fdist = nltk.FreqDist(uni_processed)
print("Top 10 unigrams: ")
print("Without stop-words removal: ")
display(sorted(uni_fdist_sw.items(), key=lambda item: item[1], reverse=True)[:10])
print("With stop-words removal: ")
display(sorted(uni_fdist.items(), key=lambda item: item[1], reverse=True)[:10])
print()

#print top 10 bigrams, trigrams, fourgrams after removing stopwords
bi_fdist_sw = nltk.FreqDist(bigrams)
bi_processed = [p for p in bigrams if (p[0] not in stopwords or p[1] not in stopwords)]
bi_fdist = nltk.FreqDist(bi_processed)
print("Top 10 bigrams: ")
print("Without stop-words removal: ")
display(sorted(bi_fdist_sw.items(), key=lambda item: item[1], reverse=True)[:10])
print("With stop-words removal: ")
display(sorted(bi_fdist.items(), key=lambda item: item[1], reverse=True)[:10])
print()

tri_fdist_sw = nltk.FreqDist(trigrams)
tri_processed = [p for p in trigrams if (p[0] not in stopwords or p[1] not in stopwords or p[2] not in stopwords)]
tri_fdist = nltk.FreqDist(tri_processed)
print("Top 10 trigrams: ")
print("Without stop-words removal: ")
display(sorted(tri_fdist_sw.items(), key=lambda item: item[1], reverse=True)[:10])
print("With stop-words removal: ")
display(sorted(tri_fdist.items(), key=lambda item: item[1], reverse=True)[:10])
print()

four_fdist_sw = nltk.FreqDist(fourgrams)
four_processed = [p for p in fourgrams if (p[0] not in stopwords or p[1] not in stopwords or p[2] not in stopwords or p[3] not in stopwords)]
four_fdist = nltk.FreqDist(four_processed)
print("Top 10 fourgrams: ")
print("Without stop-words removal: ")
display(sorted(four_fdist_sw.items(), key=lambda item: item[1], reverse=True)[:10])
print("With stop-words removal: ")
display(sorted(four_fdist.items(), key=lambda item: item[1], reverse=True)[:10])
print()
```

# Applying Smoothing


Assume additional training data in which each possible N-gram occurs exactly once and adjust estimates.

> ### $ Probability(ngram) = \frac{Count(ngram)+1}{ N\, +\, V} $

N: Total number of N-grams <br>
V: Number of unique N-grams



```python
#You are to perform Add-1 smoothing here:
#write similar code for bigram, trigram and fourgrams

# Gives k-top values for each gram, specified by input n(1<=n<=4)
def get_probdist(n, topk = None):
    # Select correct frequency dict based on n
    fdist_dict = {1: uni_fdist_sw, 2: bi_fdist_sw, 3: tri_fdist_sw, 4: four_fdist_sw}
    fdist = fdist_dict[n]
    # N = Total No. of Unigrams, gram_N = Total number of N-grams, gram_V = Number of unique N-grams possible
    N = len(uni_fdist_sw)
    gram_N = sum(fdist.values())
    gram_V = N**n
    pdist = {k: (v+1)/(gram_N+gram_V) for k, v in sorted(fdist.items(), key=lambda item: item[1], reverse=True)[:topk]}
    return pdist

# Unigrams
uni_probs_top10 = get_probdist(1, topk = 10)

# Bigrams
bi_probs_top10 = get_probdist(2, topk = 10)

# Trigrams
tri_probs_top10 = get_probdist(3, topk = 10)

# Fourgrams
four_probs_top10 = get_probdist(4, topk = 10)


#Print top 10 unigram, bigram, trigram, fourgram after smoothing
print("Top 10 Unigrams after Smoothing:")
display(uni_probs_top10)

print("Top 10 Bigrams after Smoothing:")
display(bi_probs_top10)

print("Top 10 Trigrams after Smoothing:")
display(tri_probs_top10)

print("Top 10 Fourgrams after Smoothing:")
display(four_probs_top10)
```
